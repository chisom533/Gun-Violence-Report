**DATA CLEANING AND VALIDATION**

```python
df.gender.value_counts()
```

```python
# Replace "M" with "Male" and "F" with "Female"
df['gender'] =df['gender'].replace({
    'M':'Male', 
    'F': 'Female',
    'F (identified as transgender, per police)': "Female (transgender)"})
df['gender'].value_counts()
```

```python
df.prior_signs_mental_health_issues.value_counts()
```

```python
df["prior_signs_mental_health_issues"] = df["prior_signs_mental_health_issues"].replace({
    '-':'Unclear',
    'yes':'Yes', 
    'Unknown':'Unclear',
    'Unclear ': "Unclear"
    })
df["prior_signs_mental_health_issues"].value_counts()
```

```python
df.weapons_obtained_legally.value_counts()
```

```python
#
df["weapons_obtained_legally"] = df["weapons_obtained_legally"].replace({
    "-": "Unknown", "yes": "Yes", "\nYes": "Yes",
    "Kelley passed federal criminal background checks; the US Air Force failed to provide information on his criminal history to the FBI": "Yes",
    'Yes ("some of the weapons were purchased legally and some of them may not have been")': "Yes", "Yes ": "Yes"
    })
df["weapons_obtained_legally"]. value_counts()
```

```python
df.where_obtained.value_counts()
```

```python
df['where_obtained'] = df['where_obtained'].replace({'-': "Unknown"})
df['where_obtained'].value_counts()
```

```python
df.race.value_counts()
```

```python
df['race'] = df['race'].replace({
    'white': 'White',
    'black': 'Black',
    'White ': 'White',
    '-': 'Unclear',
    'unclear': 'Unclear'
})
df['race'].value_counts()
```

```python
df["location.1"].value_counts()
```

```python
df['location.1'] = df['location.1'].replace({
    'workplace': 'Workplace',
    'Other\n': 'Other',
    'religious': 'Religious',
    '\nWorkplace': 'Workplace'
})
df['location.1'].value_counts()
```

```python
df['total_victims']=df['total_victims'].replace({"TK":10})
```

```python
df['injured']=df['injured'].replace({'(TK - "fewer than 10"': 6})
```

```python
#create new columns from df['date] into year,quarter,half,month,week
df['year'] = df['date'].dt.year
df['quarter'] = df['date'].dt.quarter
df['half'] = df['date'].dt.month.apply(lambda x: 1 if x <= 6 else 2)
df['month'] = df['date'].dt.month_name()
df['week'] = df['date'].dt.day_name()
```

**Data Visualization**

```python
#Plot Cases by year
import plotly.graph_objs as go

fig_year =go.Figure(go.Bar(
    x = df['year'].value_counts().index,
    y=df['year'].value_counts().values,
    name ='By Year'
    
))

fig_year.update_layout(
    title='Cases by year',
    xaxis_title='Year',
    yaxis_title='Number of Cases',
    height=400,
    margin=dict(l=0,r=0,t=50,b=0),
)
fig_year.show()

```

```python
grouped_by_fatalities = df.groupby('case')['fatalities'].sum()
top_10_by_fatalities = grouped_by_fatalities.sort_values(ascending=False).head(10)

import plotly.graph_objs as go

fig = go.Figure()
fig.add_trace(go.Bar(
    x=top_10_by_fatalities.values,
    y=top_10_by_fatalities.index,
    orientation='h',
    text=top_10_by_fatalities.values,
    textposition='auto', 
))

fig.update_layout(
    title='Top 10 Cases by fatalities',
    xaxis_title='Case',
    yaxis_title='Fatalities',
    height=600,
    margin=dict(l=0, r=0, t=50, b=0),
    yaxis=dict(autorange="reversed")
)

fig.show()
```

```python
#Seperate city and state in location using Split
split_col= df['location'].str.split(',',expand=True)
df['city']=split_col[0]
df['state']=split_col[1]
```

```python
#Plot Cases by States
import plotly.express as px
#Count the number of cases by State
cases_by_state = df['state'].value_counts()

#Create a bar chart
fig = px.bar( x=cases_by_state.index, y=cases_by_state.values)


#customise the layout
fig.update_layout(
    title='Cases by State',
    xaxis_title='State',
    yaxis_title='Nuber of Cases',
    margin =dict(l=0,r=0,t=50,b=0),
    height=600
)    
fig.show()
```

```python
#Plot Cases by Location.1

# count the number of cases by location
cases_by_location = df['location.1'].value_counts()

# create a line graph
fig = px.bar(x=cases_by_location.index, y=cases_by_location.values)

# customize the layout
fig.update_layout(
    title='Cases by Location',
    xaxis_title='Location',
    yaxis_title='Number of Cases',
    height=600,
    margin=dict(l=0, r=0, t=50, b=0),
    xaxis_tickangle=-45,
)

# show the plot
fig.show()
```

```python
#Plot Cases By Gender
#count the number of cases by gender 
cases_by_gender = df['gender'].value_counts()

#create a bar chart
fig= px.bar(x=cases_by_gender.index, y=cases_by_gender.values)

#customise the layout
fig.update_layout(
   title='Cases by Gender',
   xaxis_title='Gender',
   yaxis_title='Number of Cases',
   height=600,
   margin=dict(l=0, r=0, t=50, b=0),
)
fig.show()
```

```python
#Define age_ranges and age_labels
df['age_of_shooter'] = df['age_of_shooter'].replace({'-':0})
df['age_of_shooter'] =df['age_of_shooter'].astype(int)
age_ranges =[0,10,19,40,60,120]
age_labels =['Child', 'Teenage', 'Young Adult', 'Middle-aged Adult', 'Old-aged Adult']

#Categorize age group using cut() funtions
df['age_group'] = pd.cut(df['age_of_shooter'], bins =age_ranges, labels=age_labels)
df['age_group'] .value_counts()
```

```python
#Plot Cases By age_group
#count the number of cases by age_group
cases_by_age_group = df['age_group'].value_counts()

#create a bar chart
fig= px.bar(x=cases_by_age_group.index, y=cases_by_age_group.values)

#customise the layout
fig.update_layout(
   title='Cases by Age_Group',
   xaxis_title='age_group',
   yaxis_title='Number of Cases',
   height=600,
   margin=dict(l=0, r=0, t=50, b=0),
)
fig.show()
```
